# HANDLE MISSING VALUES

Identify missing values

```python
df.isnull().sum()
```

Handle missing values

```python
df = df.dropna()
```

### Checking for Missing Values

The `isnull().sum()` function is used to identify missing values in each column of the dataset.  
This helps determine which columns require cleaning before further analysis.

### Handling Missing Values

Rows containing missing values were removed using the `dropna()` method.  
This ensures the dataset is clean and suitable for basic analysis.

```python
df.isnull().sum()
```

### Verifying Missing Values Removal

After cleaning, the dataset was checked again to confirm that no missing values remain.

# REMOVE DUPLICATE ROWS

### Check for duplicate rows

```python
df.duplicated().sum()
```

### Remove duplicate rows

```python
df = df.drop_duplicates()
```

# CREATE DERIVED COLUMN (total_amount)

```python
df['total_amount'] = df['initial_price'] * df['ratings_count']
```

### Verify the new column

```python
df[['initial_price', 'ratings_count', 'total_amount']].head()
```

### Creating a Derived Column

A new column named `total_amount` was created using numeric columns from the dataset.
The `initial_price` and `ratings_count` columns were used to perform a mathematical
operation and generate a derived feature for analysis.
